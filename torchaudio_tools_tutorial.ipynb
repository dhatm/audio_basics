# 4.0 RESAMPLING

[back to table of contents](#table_of_contents)

Resampling includes:
- Upsampling (with sinc interpolation to fill in missing points)
- Downsampling (with lowpass filtering to avoid aliasing)

Note: you can use torchaudio.transforms.Resample or torchaudio.functional.resample
- transforms.Resample -- precomputes and caches the kernel used for resampling
- functional.resample -- computes it on the fly

Thus using transforms.Resample will result in a speedup when resampling multiple waveforms using the same parameters



## 4.1 Upsampling

Upsample an audio waveform from one frequency to a higher one.  
Default is to use sinc interpolation (sinc_interp_hann) to fill in the missing points.

### Imports

```python
import torchaudio.transforms as T
import torchaudio.functional as F
```

### Waveform and (re)sample rates

```python
waveform, sample_rate = torchaudio.load(SAMPLE_WAV)

resample_rate = sample_rate * 2

print(f"sample_rate: {sample_rate}, resample_rate: {resample_rate}")
```

### Do Upsampling

```python
resampler = T.Resample(sample_rate, resample_rate, dtype=waveform.dtype) #computes & saves resampling kernal
resampled_waveform1 = resampler(waveform) #apply saved resampling kernal

resampled_waveform2 = F.resample(waveform, sample_rate, resample_rate) #computes kernal on the fly
```

### Check Results

```python
#check audio -- should all sound the same
play_audio(waveform, sample_rate)
play_audio(resampled_waveform1, resample_rate)
play_audio(resampled_waveform2, resample_rate)
```

```python
# check waveforms -- should look similar (just more points that can't easily be seen)
plot_waveform(waveform, sample_rate,title='original waveform',xlabel='time (s)',ylabel='amplitude')
plot_waveform(resampled_waveform1, resample_rate,title='resampled1 waveform',xlabel='time (S)',ylabel='amplitude')
plot_waveform(resampled_waveform2, resample_rate,title='resampled2 waveform',xlabel='time (s)',ylabel='amplitude')
```

```python
# check spectrograms -- everything below original Nyquist frequencey (sample_rate/2 = 8kHz) should be the same
plot_specgram(waveform, sample_rate, title='Spec Original', xlabel='time (s)', ylabel='frequency (Hz)')
plot_specgram(resampled_waveform1, resample_rate, title='Spec Resampled1') #freq should go twice as high (16kHz)
plot_specgram(resampled_waveform2, resample_rate, title='Spec Resampled2') #freq should go to 16kHz

```

## 4.2 Downsampling

Downsample an audio waveform from one frequency to a lower one. We (usually) need to lowpass filter here to 
avoid aliasing. Default is to use Hann window filter (sinc_interp_hann) 

```python
import torchaudio.transforms as T
import torchaudio.functional as F
```

### Waveform and (re)sample rates

```python
waveform, sample_rate = torchaudio.load(SAMPLE_WAV)

resample_rate = sample_rate * 0.5

print(f"sample_rate: {sample_rate}, resample_rate: {resample_rate}")
```

### Filter and Downsample

<br>

<ins>Parameters:</ins>
- **Lowpass filter width**: 
   Larger lowpass_filter_width --> sharper, more precise filter, but more computationally expensive.
- **Rolloff**:
   Lower rolloff value --> more attenuation (& thus reduced aliasing) near Nyquist. This will also reduce higher frequencies we may want to keep so beware.
- **Window filter function**:
   - default is Hann, which is a weighted cosine function. 
   - Kaiser is better. This is a near optimal window function that contains an additional ``beta`` parameter that allows for the design of the smoothness of the filter and width of impulse. This can be controlled using the ``resampling_method`` parameter.
   - Blackman or Parks-McClellan is more precise but can be overkill (to do)
   
   

```python
resampled_waveform = F.resample(waveform, sample_rate, resample_rate, lowpass_filter_width=12, 
                                rolloff=0.90, resampling_method="sinc_interp_kaiser")

```

### Check Results

```python
waveform8khz, sample_rate_8khz = torchaudio.load(SAMPLE_WAV_8000) #loaded from url, for comparision here
```

```python
#check audio -- downsampled version should have less higher frequency content
play_audio(waveform, sample_rate)
play_audio(resampled_waveform, resample_rate)#should have less high freq content than previous, & sound like below
play_audio(waveform8khz, sample_rate_8khz)
```

```python
# check waveforms -- may look similar (just less points that can't easily be seen) or may look slightly smoother
plot_waveform(waveform, sample_rate,title='original waveform',xlabel='time (s)',ylabel='amplitude')
plot_waveform(resampled_waveform, resample_rate,title='resampled waveform',xlabel='time (s)',ylabel='amplitude')
plot_waveform(waveform8khz, resample_rate,title='waveform 8kHz',xlabel='time (s)',ylabel='amplitude')

```

```python
# check spectrograms -- everything below original Nyquist frequencey * 0.5( = 4kHz) should be similar 
# if aliasing is present, it won't look the same
plot_specgram(waveform, sample_rate, title='Spec Original')
plot_specgram(resampled_waveform, resample_rate, title='Spec Downsampled')
plot_specgram(waveform8khz, sample_rate_8khz, title='Spec Waveform 8kHz')


print(f"sample_rate: {sample_rate}, resample_rate: {resample_rate}, sample_rate_8khz: {sample_rate_8khz}")
```
